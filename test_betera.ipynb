Условия задачи

Выгрузить игроков, которые:     
С 12:00 14.03.2022 сделали ставку в пре-матч на события раздела «Киберспорт».
Минимальная сумма ставки для участия в Бонусном предложении составляет 10 BYN с коэффициентом не менее 1,5, в экспрессе все события на киберспорт с кф каждого события от 1.50
Ставка должна быть рассчитана не позднее 12:00 15.03.2022.
Ставки вида «система» не учитываются.
Ставки CashOut, возвраты и FreeBet не участвуют в Бонусе.

Решение необходимо в двух вариантах: 
SQL 
Python

## Решение на python

```python
import pandas as pd
import numpy as np
```

```python
# Загрузим датасеты

events_df = pd.read_csv('events.csv')
bets_df = pd.read_csv('bets.csv')
```

```python
# Преобразуем данные в столбцах в данные типа дата для удобства дальнейшей работы

bets_df['create_time'] = pd.to_datetime(bets_df['create_time'])
bets_df['settlement_time'] = pd.to_datetime(bets_df['settlement_time'])
```

```python
# Отфильтруем датасет с данными о ставках по условию

new_df = bets_df[(bets_df['create_time'] >= '2022-03-14 12:00:00') &              # время ставки с 12:00 14.03.2022
        (bets_df['event_stage'] == 'Prematch') &                                  # ставки в пре-матч
        (bets_df['amount'] >= 10) &                                               # минимальная сумма ставки 10 BYN
        ((bets_df['accepted_odd'] >= 1.5) |                                        # коэффициент не менее 1,5    
        ((bets_df['bet_type'] == 'Express') & (bets_df['accepted_odd'] > 1.50))) & # в экспрессе коэфициент от 1,5
        (bets_df['settlement_time'] <= '2022-03-15 12:00:00') &                   # время рассчета ставки не позднее 12:00 15.03.2022
        (bets_df['bet_type'] != 'System') &                                       # ставка вида не "система"
        ( ~ bets_df['result'].isin(['Cashout', 'Return', 'TechnicalReturn'])) &   # не Cashout и не возвраты
        (bets_df['is_free_bet'] == False)                                         # не FreeBet
       ]
new_df
```

```python
# Выберем только ставки на события раздела «Киберспорт»

res_df = new_df.merge(events_df[events_df['sport'] == 'E-Sports'], how = 'inner', on = 'event_id')
res_df
```

## Решение на SQL

```python
# Импортируем библиотеку sqlite3 для работы с sql

import sqlite3

conn = sqlite3.connect('test.db')

conn.commit()
```

```python
# Создаем тестовую БД

conn = sqlite3.connect("test.db")

# Создаем курсор

cur = conn.cursor()

# Загружаем датафрейм bets_df в БД

bets_df.to_sql("bets", conn, if_exists='replace', index=False)
```

```python
# Загружаем датафрейм events_df в БД

events_df.to_sql("events", conn, if_exists='replace', index=False)
```

```python
# Фильтруем данные на основании данного в задаче условия

res_df = pd.read_sql_query( ''' 
                   SELECT * FROM bets,events
                   
                   WHERE bets.event_id = events.event_id
                   AND bets.create_time >= '2022-03-14 12:00:00' 
                   AND bets.event_stage == 'Prematch'
                   AND bets.amount >= 10
                   AND (bets.accepted_odd >= 1.5
                   OR (bets.bet_type == 'Express' AND bets.accepted_odd > 1.50))
                   AND bets.settlement_time <= '2022-03-15 12:00:00'
                   AND bets.bet_type != 'System'
                   AND bets.result != 'Cashout' AND result != 'Return' AND result != 'TechnicalReturn'
                   AND bets.is_free_bet == False
                   AND events.sport = 'E-Sports'
                   
                   
               ''', conn)
res_df
```

```python
# Закрываем соединение

conn.close()
```
